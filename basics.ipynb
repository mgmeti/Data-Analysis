```python
import pandas as pd
import numpy as np
```

DataFrame for given dataset

```python
data = {'Name':['P. Lee','R. Jones','J. Smith','A. Patel','M. Owen','S. Green','N. Cook','W. Hands','P. Rice','F. Marsh'],
        'Age':[35, 52, 45, 70, 24, 43, 68, 77, 45, 28 ],
        'Gender': ['Female',  'Male', 'Male', 'Female', 'Male', 'Male', 'Male','Female', 'Female', 'Male'],
        'Blood Group':['ARh+', 'ORh-', 'ORh+', 'ORh-', 'ARh-', 'ORh-', 'ARh+', 'ORh-', 'ORh+', 'ORh+'],
        'Weight (kg)': [50, 115, 96, 41, 79, 109, 73, 104, 64, 136],
        'Height (m)':[1.52, 1.77, 1.83, 1.55, 1.82, 1.89, 1.76, 1.71, 1.74, 1.78],
        'Systolic Blood Pressure (mmHg)':[68, 110, 88, 76, 65, 114, 108, 107, 101, 121],
        'Diastolic Bllod Pressure (mmHg)':[112, 154, 136, 125, 105, 159, 136, 145, 132, 165],
        'Diabetes': [0, 1, 0, 0, 0, 1, 0, 1, 0, 1]}
```

```python
# Create DataFrame
df = pd.DataFrame(data)
df.head()
```

a) min-max normalisation

```python
# Function to normalize weight
def normalize_weight(weight):
    min_weight = df['Weight (kg)'].min()
    max_weight = df['Weight (kg)'].max()
    normalized_weight = (weight - min_weight) / (max_weight - min_weight)
    return round(normalized_weight, 3)

df['Normalized Weight'] = df['Weight (kg)'].apply(normalize_weight)
df.head(len(df))
```

b) Binned weight
<br>
The Weight (kg) variable into three categories:
low (less than 60 kg), medium (60-100 kg), and high (greater than 100 kg).

```python
# Function to bin weight
def bin_weight(weight):
    if weight < 60:
        return 'Low'
    elif weight <= 100:
        return 'Medium'
    else:
        return 'High'

df['Binned Weight'] = df['Weight (kg)'].apply(bin_weight)
df.head(len(df))
```

c) Body mass index using formula

```python
# Function to calculate BMI
def calculate_bmi(row):
    weight = row['Weight (kg)']
    height = row['Height (m)']
    bmi = weight / (height ** 2)
    return bmi

df['BMI'] = df.apply(calculate_bmi, axis=1)
df.head(len(df))
```

### Q2

Create
a DataFrame from the data.

```python

data = {
    'x': [92, 145, 30, 70, 75, 105, 110, 108, 45, 50, 160, 155, 180, 190, 63, 85, 130, 132],
    'y': [6.3, 7.8, 3.0, 5.5, 6.5, 5.5, 6.5, 8.0, 4.0, 5.0, 7.5, 9.0, 8.6, 10.0, 4.2, 4.9, 6, 7]
}

df = pd.DataFrame(data)
df.head(len(df))
```

```python
# Calculate the means
mean_x = df['x'].mean()
mean_y = df['y'].mean()

# Calculate the deviations from the means
df['x - x_bar'] = df['x'] - mean_x
df['y - y_bar'] = df['y'] - mean_y

# Calculate the product of deviations
df['(x - x_bar)(y - y_bar)'] = df['x - x_bar'] * df['y - y_bar']
df.head(len(df))
```

The correlation coefficient

```python
# Compute the correlation coefficient (r)
numerator = df['(x - x_bar)(y - y_bar)'].sum()
denominator = (len(df) - 1) * np.std(df['x']) * np.std(df['y'])
r = numerator / denominator

# Display the correlation coefficient (r)
print(f"Correlation coefficient (r): {r}")
```

```python
# Determine if x and y are correlated
if abs(r) > 0.5:
    print("The attributes x and y are correlated.")
else:
    print("The attributes x and y are not correlated.")
```

The correlation coefficient (r) between variables x and y was calculated to be 0.96. <br>This indicates a strong positive correlation between the two variables. The high value of r suggests that as the values of x increase, the values of y also tend to increase consistently. <br>Thus, there is a significant and positive linear relationship between x and y in the given dataset.
